```python
import pandas as pd

df = pd.read_csv("./filtered/compaction_sequential_filtered.csv")
df
```

```python
# Sequential Range Query W/ Compaction

import matplotlib.pyplot as plt

df.plot(kind='bar', x='Selectivity', y=['write file count'], figsize=(10, 6), rot=290)
plt.title('Sequential Range Query with Compaction')
plt.xlabel('Selectivity')
plt.ylabel('Write I/O')
plt.show()
```

```python
df2 = pd.read_csv("./filtered/compaction_nonsequential_Filter.csv")
df2
```
